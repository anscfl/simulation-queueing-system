```python
from math import pi, sqrt
import random
def monte_carlo(n, R):
    point_counter = 0
    for i in range(n):
        x = random.uniform(0, R)
        y = random.uniform(0, 2)
        if y <= (4*x*sqrt(R ** 2 - x ** 2)):
            point_counter += 1
    return (2*R)**3 * point_counter / n

R = 1
N = 10000
print(f"N = {N}, monte_carlo = {monte_carlo(N, R)}")
print(f'Точное значение = {16*R**3 / 3}')
```

```python
from random import randint
from math import sqrt

summ = 0
i = 0
cnt = 0
x = [0, 0, 0, 0]

while (i < 10000):
    r = randint(0,3)
    x[r] += 1
    i += 1
    if ((x[0] > 2) or (x[1] > 2) or (x[2] > 2) or (x[3] > 2)):
        summ += x[0] + x[1] + x[2] + x[3]
        cnt += 1;
        for k in range(4):
            x[k] = 0

EX = summ / cnt
print(f"Мат. ожидание = {EX}")
s = sum([(x[k] - EX) ** 2 for k in range(4)]) / (cnt - 1)
print(f"Среднеквадр. отклонение = {sqrt(s)}")
```

```python
from math import log 
from random import random
import numpy as np

t_act_source = 0 #Момент активации источника
t_modeling = 100 #Длительность моделирования
t_now = 0 #Текущий момент модельного времени
lambd = 5#Интенсивность входящего потока требований
mu = 1 #Интенсивность обслуживания требования прибором
employment_sign = 0 #Признак занятости прибора
Q = [0]*5000
N = 0 #Число требований в системе
B = 0 #Кол-во мест в очереди
rnd = random()
t_act_device = 0 #Момент активации прибора
moments_start = []
moments_end = []
moments = []
while t_now < t_modeling:
    indicator = 0
    
    #Формирование требования
    if (t_act_source == t_now and B < 4):
        print(f"Момент формирования требования: {t_now}")
        indicator = 1
        N += 1
        B += 1
        Q[N] = t_now
        t_act_source = t_now - log(rnd) / lambd
    #Обслуживание
    if (employment_sign == 0 and B > 0):
        print(f"Момент начала обслуживания: {t_now}")
        indicator = 1
        employment_sign = 1
        t_act_device = t_now - log(rnd) / mu
        B -= 1
        tmp = Q[N]
    if (t_act_device == t_now or B == 4):
        print(f"Момент завершения обслуживания: {t_now}")
        indicator = 1
        employment_sign = 0
        t_act_device = t_modeling + 0.000001
        print(f"Момент поступления: {Q[1]}\nМомент выхода из СМО: {t_now}")
        moments_start.append(tmp)
        moments_end.append(t_now)
        for j in range(N):
            Q[j] = Q[j+1]
        N -= 1
    if (indicator == 0):
        if (t_act_device < t_act_source):
            t_now = t_act_device
        else:
            t_now = t_act_source

for i in range(0, len(moments_start)):
    moments.append(moments_end[i] - moments_start[i])
print(f"М.о. длительности пребывания требований в системе = {np.mean(moments)}")
print(f"М.о. числа требований в системе = {np.mean(moments) * lambd}")
```
